## House Prices Prediction Using Gradient Boosting

Ask a home buyer to describe their dream house, and they probably won't begin with the height of the basement ceiling or the proximity to an east-west railroad. But this playground competition's dataset proves that much more influences price negotiations than the number of bedrooms or a white-picket fence.

With 79 explanatory variables describing (almost) every aspect of residential homes in Ames, Iowa, this competition challenges you to predict the final price of each home.

### 10. Data Preparation for Modelling

```python
X_train = df_train.drop('SalePrice', axis = 1)
X_train.shape
```

```python
y_train = df_train['SalePrice']
y_train.shape
```

```python
X_test = df_test
X_test.head()
```

```python
train_len = len(X_train)

df = pd.concat([X_train, X_test], ignore_index=True)
df.head()
```

```python
# Checking shape of df
df.shape
```

```python
# Checking column information
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder

Encoder = LabelEncoder()
categorical = df.select_dtypes(include=['object'])

for column in categorical:
    df[column] = Encoder.fit_transform(df[column].astype('str'))

X_train = df[:train_len]
X_test = df[train_len:]
```

```python
X_train.head()
```

```python
X_test.head()
```

### 11. Model building

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import cross_val_score
```

```python
GBR = GradientBoostingRegressor(loss='huber', n_estimators=2000, learning_rate=.05)

cv_score = cross_val_score(GBR, X_train, y_train, cv=10, n_jobs=-1)

print(f"GBR Model Max Test Score: {cv_score.max().round(4)*100}%")
```

```python
GBR.fit(X_train, y_train)

print(f"GBR Model Train Score: {GBR.score(X_train, y_train).round(4)*100}%")
```

### 12. Prediction on test dataset

```python
y_pred = GBR.predict(X_test)
```

```python
y_pred
```

```python
submission = pd.DataFrame({'Id': X_test['Id'], 'SalePrice' : y_pred})
```

```python
submission.to_csv("Pred_SalePrice.csv", index= False)
```
